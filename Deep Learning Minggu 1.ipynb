# Tipe data python
## List
List adalah struktur data pada python yang mampu menyimpan lebih dari satu data(value) dan datanya(valuenya) bisa diubah-ubah, seperti array. Nilai-nilai itu disimpan kedalam sebuah index.
List umumnya ditandai dengan kurung 
[... , ..., ...].


```python
a = ['a', 'b', 'c', 'd']
a[2]
```

```python
a[3] = 'e'
a[3]
```

```python
del a[3]
a
```

# Tipe Data Python
## Tuple
Tuple sebenarnya hampir sama dengan tipe data list. Yang membedakannya adalah nilai/data dari tuple tidak dapat diubah-ubah seperti tipe data list.
Tuple umumnya ditandai dengan kurung (..., ..., ...)


```python
b = ('a','b','c')
b[2]
```

```python
b[2] = 'd'
```

# Tipe Data Python
## Dictionary
Dictionary adalah struktur data pada python yang berisi tentang key dan value.
Untuk melihat sebuah value dari dictionary diharuskan mengetahui key nya terlebih dahulu.
Dictionary umumnya ditandai dengan kurung kurawal {key : value, key : value, key : value}

```python
a = {'nama': 'irfan', 'kelas': '4IA14', 'umur':17 , 'npm':'55415187'}
a['umur']
```

```python
del a['npm']
a
```

```python
a['npm'] = '55415187'
a
```

# Perulangan Python
## for loop
## while loop

```python
ulang = 10
for i in range (ulang):
    print('perulangan ke- '+str(i))
```

```python
for a in range (5):
    for b in range (a+1):
        print('*',end='')
    print()
```

```python
i = 0
while(i<5):
    print(i)
    i+=1
```

```python
kosong = ""
baris = 1
while(baris < 10):
    kolom = baris
    while(kolom > 0):
        kosong = kosong + " * "
        kolom-=1
    kosong = kosong +"\n"
    baris+=1
print(kosong)
```

```python
kosong = ""
baris = 10
while(baris >= 0):
    kolom = baris
    while(kolom > 0 ):
        kosong = kosong + " * "
        kolom-=1
    kosong = kosong + "\n" 
    baris-=1
print(kosong)
```
